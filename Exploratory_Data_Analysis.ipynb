>Type of the coasters made by the manufacture

```python
def func(x):
    if pd.notnull(x) and isinstance(x, str) and x.startswith("Arrow"):
        return "Arrow"
    elif pd.notnull(x) and isinstance(x, str) and x.startswith("Vekoma"):
        return "vekoma"
    return x

df["Manufacturer"] = df["Manufacturer"].apply(func)
df["Manufacturer"].unique()
```

>insights into the distribution of manufacturers based on the materials they primarily use for coaster making

```python
plt.figure(figsize=(16,8))
sns.countplot(data=df,x="Manufacturer",hue="Type_Main",palette="flare")
plt.xticks(rotation=90)
plt.show()
```

>Top 10 Manufacturers

```python
df['Manufacturer'].value_counts()
```

```python
df2 = df.groupby('Manufacturer')['Speed_mph']\
.agg(['mean','median','count'])\
.query('count<=10')\
.sort_values('count',ascending=False).head(10)
df2
```

```python
df2.plot(kind='bar',figsize=(16,8),title='Top 10 Manufacturers',colormap="flare")
plt.show()
```

>What are the locations with the faster roller coasters(minimum 10)

```python
df['Location'].value_counts()
```

```python
df.query('Location !="Other"') \
.groupby('Location')['Speed_mph'] \
.agg(['mean','median','count']) \
.query('count >= 10') \
.sort_values('mean',ascending=False) \
.plot(kind='bar',figsize=(16,8),title = 'Locations with the fastest roller coasters')
plt.show()
```
